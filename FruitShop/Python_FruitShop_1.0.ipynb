水果类： 名称、价格、数量

*************************************水果超市********************************
1.   查询全部水果
2.   增加水果（上架）
3.   删除水果  （下架）
4.   修改水果信息 （价格、数量）
5.   排序（价格）
6.   查询指定水果
7.   退出系统
*******************************************************************************

操作：
选择1，调用 1.查询全部水果，呈现效果
选择2，调用 2 增加水果   ： name , price , count

```python
import os
class Fruit:
    def __init__(self, name, price, quantity):
        self.name = name
        self.price = price
        self.quantity = quantity


class Fruit_Manager:
    def __init__(self):
        self.fruits = []

    # 添加水果
    def add_fruit(self, fruit):
        self.fruits.append(fruit)

    # 删除水果
    def remove_fruit(self, fruit_name):
        for fruit in self.fruits:
            if fruit.name == fruit_name:
                self.fruits.remove(fruit)
                return True
        return False

    # 更新水果信息
    def update_fruit(self, fruit_name, new_price, new_quantity):
        for fruit in self.fruits:
            if fruit.name == fruit_name:
                fruit.price = new_price
                fruit.quantity = new_quantity
                return True
        return False

    # 获取所有水果信息
    def get_all_fruits(self):
        return self.fruits
    
    # 获取单个水果信息
    def get_fruit(self, fruit_name):
        for fruit in self.fruits:
            if fruit.name == fruit_name:
                return fruit
        return None

    # 按价格排序
    def sort_by_price(self):
        self.fruits.sort(key=lambda fruit: fruit.price)


# 测试
if __name__ == '__main__':
    fruit_manager = Fruit_Manager()
    while True:
        print(f"*************************************水果超市********************************")
        print("1.查询全部水果")
        print("2.添加水果")
        print("3.删除水果")
        print("4.更新水果信息")
        print("5.按价格排序")
        print("6.查询指定水果")
        print("7.退出系统")
        print("*******************************************************************************")
        choice = input("请输入您的选择：")
        if choice == "1":
            #fruit_manager = Fruit_Manager()
            print(f"==========查询全部水果信息==========")
            fruits = fruit_manager.get_all_fruits()
            if len(fruits) == 0:
                print("水果超市里还没有水果")
            else:
                for fruit in fruits:
                    print(f"{fruit.name}：{fruit.price}元，{fruit.quantity}个")
        elif choice == "2":
            #fruit_manager = Fruit_Manager()
            print(f"==========上架水果==========")
            name = input("请输入水果名称：")
            price = float(input("请输入水果价格："))
            quantity = int(input("请输入水果数量："))
            fruit = Fruit(name, price, quantity)
            fruit_manager.add_fruit(fruit)
            print(f"{name}水果添加成功")
        elif choice == "3":
            #fruit_manager = Fruit_Manager()
            print(f"==========下架水果==========")
            name = input("请输入要删除的水果名称：")
            if fruit_manager.remove_fruit(name):
                print(f"{name}水果删除成功")
            else:
                print(f"{name}水果不存在")
        elif choice == "4":
            #fruit_manager = Fruit_Manager()
            print(f"==========更新水果信息==========")
            name = input("请输入要更新的水果名称：")
            new_price = float(input("请输入新的水果价格："))
            new_quantity = int(input("请输入新的水果数量："))
            if fruit_manager.update_fruit(name, new_price, new_quantity):
                print(f"{name}水果信息更新成功")
            else:
                print(f"{name}水果不存在")
        elif choice == "5":
            #fruit_manager = Fruit_Manager()
            print(f"==========按价格升序排列==========")
            fruit_manager.sort_by_price()
            fruits = fruit_manager.get_all_fruits()
            if len(fruits) == 0:
                print("水果超市里还没有水果")
            else:
                print("按照价格排序后的水果信息：")
                for fruit in fruits:
                    print(f"{fruit.name}：{fruit.price}元，{fruit.quantity}个")
        elif choice == "6":
            #fruit_manager = Fruit_Manager()
            print(f"==========查询指定水果==========")
            name = input("请输入要查询的水果名称：")
            fruit = fruit_manager.get_fruit(name)
            if fruit is None:
                print(f"{name}水果不存在")
            else:
                print(f"{name}水果信息：{fruit.price}元，{fruit.quantity}个")
        elif choice == "7":
            os._exit(0)
        else:
            print("输入错误，请重新输入")
```
