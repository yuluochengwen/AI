#### PAT-1063 计算谱半径

```python
import math

n = int(input())
max_radius = 0.0

for _ in range(n):
    a, b = map(int, input().split())
    radius = math.sqrt(a**2 + b**2)
    if radius > max_radius:
        max_radius = radius

print("{:.2f}".format(max_radius))
```

#### PAT-1014 福尔摩斯的约会
1. 第一二行中，第一对相同的大写字母代表周几(第几个字母就代表周几)
2. 第一二行中，第二对相同的大写字母代表几点(0~9,A~N)
3. 第三四行中，第一对相同的字母(无法确定大小写是否有区别,但题例中是小写(后续证明，只要字母相对就行，大写=大写，小写=小写))在第几位(0开始计数)就代表第几分钟

```python
str1 = input()
str2 = input()
str3 = input()
str4 = input()
count = 0
result = []
day_dict = {'A':'MON','B':'TUE','C':'WED','D':'THU','E':'FRI','F':'SAT','G':'SUN'}
HH_dict = {'0':'00','1':'01','2':'02','3':'03','4':'04','5':'05','6':'06','7':'07','8':'08','9':'09',
           'A':'10','B':'11','C':'12','D':'13','E':'14','F':'15','G':'16','H':'17','I':'18','J':'19',
           'K':'20','L':'21','M':'22','N':'23'}
for i in range(min(len(str1), len(str2))):
    if str1[i] == str2[i]: # 相同的字符
        #if 'A' <= str1[i] <= 'G' and count == 0:
        if ord(str1[i]) in range(65,72) and count == 0: # 大写字母,切是第一次出现
            result.append(day_dict[str1[i]])
            count += 1
        elif str1[i] in HH_dict and count == 1: # 时间,切是第二次出现相同
            result.append(HH_dict[str1[i]])
            break

for i in range(min(len(str3), len(str4))):
    if str3[i] == str4[i] and str3[i].isalpha(): # 相同的字母
        result.append(f'{i:02d}')
        break

print(f'{result[0]} {result[1]}:{result[2]}')

```

```python
str1 = input()
str2 = input()
str3 = input()
str4 = input()

result = []
day_dict = {'A': 'MON', 'B': 'TUE', 'C': 'WED', 'D': 'THU', 
            'E': 'FRI', 'F': 'SAT', 'G': 'SUN'}
HH_dict = {'0': '00', '1': '01', '2': '02', '3': '03', '4': '04',
           '5': '05', '6': '06', '7': '07', '8': '08', '9': '09',
           'A': '10', 'B': '11', 'C': '12', 'D': '13', 'E': '14',
           'F': '15', 'G': '16', 'H': '17', 'I': '18', 'J': '19',
           'K': '20', 'L': '21', 'M': '22', 'N': '23'}

# 查找日期和小时
found_day = False
for i in range(min(len(str1), len(str2))):
    if str1[i] == str2[i]:
        if not found_day and 'A' <= str1[i] <= 'G':  # 第一对相同大写字母（A-G）
            result.append(day_dict[str1[i]])
            found_day = True
        elif found_day and str1[i] in HH_dict:  # 第二对相同字符（0-9或A-N）
            result.append(HH_dict[str1[i]])
            break

# 查找分钟
for i in range(min(len(str3), len(str4))):
    if str3[i] == str4[i] and str3[i].isalpha():  # 相同字母（大小写均可）
        result.append(f"{i:02d}")  # 格式化为两位数
        break

# 输出结果
print(f"{result[0]} {result[1]}:{result[2]}")
```
