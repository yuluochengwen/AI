#### 存储过程

1. 存储过程是一种数据库对象，它是一组预编译的SQL语句，存储在数据库中，可以被调用执行。存储过程可以减少网络流量、提高数据库性能、提高代码的可重用性。

````python
-- 例1
-- 第一步：设置分隔符（避免分号冲突）
DELIMITER //
-- 第二步：创建存储过程
CREATE PROCEDURE mycount(OUT param INT)
BEGIN
    SELECT COUNT(*) INTO param FROM fruit;
END //
-- 第三步：恢复默认分隔符
DELIMITER ;
-- 第四步：调用存储过程（在新查询中执行或确保有分隔符）
CALL mycount(@param);
-- 第五步：查看结果
SELECT @param AS fruit_count;


-- 例2
create procedure mycount2(in param int)
begin
	declare num int;
	set num = 0;
	select count(*) into num from fruit;
	set num = num + param;
	select num;
end;
call mycount2(100);


-- 例3
create procedure mycount2(in param int)
begin
	declare num int;
	set num = 0;
	select count(*) into num from fruit;
	set num = num + param;
	select num;
end;
call mycount2(100);


-- 例4
create procedure mycount3(in type int)
begin
	declare c varchar(50);
	if type = 0 then set c = 'value is 0';
	elseif type = 1 then set c = 'value is 1';
	elseif type = 2 then set c = 'value is 2';
	else set c = 'validate value';
	end if;		#终止if语句
	select c;
end;
call mycount3(1);


-- 例5
create procedure myprocwhen(in type int)
BEGIN
	declare c varchar(50);
	case type
	when 0 then set c = 'value is 0';
	when 1 then set c = 'value is 1';
	else set c = 'validate value';
	end case;
	select c;
END;
call myprocwhen(10);


-- 例6
DROP PROCEDURE IF EXISTS myprocwhile;
create procedure myprocwhile(in n int)
BEGIN
	declare i int default 0;
	declare s int default 0;
	
	while i <= n do 
		set s = s + i;
		set i = i + 1;
	end while;
	select s as total_sum;
END;
call myprocwhile(10);


-- 例7
DROP PROCEDURE IF EXISTS myprocrepeat;
create procedure myprocrepeat(in n int)
BEGIN
	declare i int default 0;
	declare s int default 0;
	repeat 
		set s = s + i;
		set i = i + 1;
	until i > n end repeat;
	select s;
END;
call myprocrepeat(10);


-- 例8
DROP PROCEDURE IF EXISTS myprocloop;
create procedure myprocloop(in n int)
BEGIN
	declare i int default 0;
	declare s int default 0;
	loop_label:
		loop
			set s = s + i;
			set i = i + 1;
			if i > n then leave loop_label;
			end if;
		end loop;
	select s;
END;
call myprocloop(10);
````

#### 触发器
1. 触发器的定义：触发器是一种特殊的存储过程，它在特定事件发生时自动执行。
2. 触发器的作用：触发器可以帮助我们在数据表上执行一些自动化的操作，比如在数据表中插入、更新或删除数据时自动执行一些操作，比如对数据进行审计、数据完整性检查等。
3. 触发器的创建：创建触发器的语法如下：
```
CREATE TRIGGER trigger_name
BEFORE/AFTER INSERT/UPDATE/DELETE ON table_name
FOR EACH ROW
BEGIN
   -- trigger body
END;
```
4. 触发器的类型：触发器分为三种类型：
- 行级触发器：在每行记录上触发，可以对每行记录进行操作。
- 语句级触发器：在SQL语句执行时触发，可以对整个SQL语句进行操作。
- 数据库级触发器：在整个数据库上触发，可以对整个数据库进行操作。
5. 触发器的执行顺序：触发器的执行顺序是先执行行级触发器，然后执行语句级触发器，最后执行数据库级触发器。


````python
-- 插入类型的触发器
create trigger my_insert
after insert on t_user
for each row 
begin
	insert into t_user2(id,name,address) 
	values (new.id,new.name,new.address);
end;


-- 更新类型的触发器
create trigger my_uppdate
after update on t_user
for each row 
begin 
	update t_user2 set name = new.name where id = old.id;
end;

-- 删除类型的触发器
create trigger my_delect
after delete on t_user
for each row 
begin 
	delete from t_user2 where id = old.id;
end;
````
