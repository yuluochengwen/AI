### 第三章 DataFrame的排序和筛选

### 1.排序（DataFrame的方法）

#### (1) DataFrame按列值重新排序(sort_values())

```python
#DataFrame的方法sort_values（），对整个DataFrame按列值排序，不能按行值排序。
#此方法也对Series能用，因为Series也是一种特殊的DateFrame
import pandas as pd
oly=pd.read_csv('olympics.csv',skiprows=4)
#按照oly中的'Edition'和'Athlete'对oly中的数据进行重新排序。
#inplace默认为False,不改变oly;若inplace=True,则oly改变为重新排序后的DataFrame
oly1=oly.sort_values(by=['Edition','Athlete'],ascending=True,inplace=False) 
oly1.head(4)
```

#### (2) 按照行索引名称和列索引名称排序(sort_index())

```python
import pandas as pd
oly1=oly.sort_index(axis=1,ascending=False)  #按照列标签名字排序的方式显示
oly1.head(2)
oly2=oly.sort_index(axis=0,ascending=False)  #按照行标签名字排序的方式显示
oly2.head(2)
```

### 2.DataFrame数据的筛选（.ix()已不建议使用）

#### (1)位置索引筛选(重点掌握)

```python
import numpy as np
import pandas as pd
oly=pd.read_csv('olympics.csv',skiprows=4)
#1 直接筛选
oly[3:5]   #只能写成切片形式，不能写oly[2]直接选择一行,可通过oly[2:3]实现
oly['City']  #和oly.City效果一样，不可以加行名或行号。
oly[['City','Sport','Event']]  #切出一个DataFrame
#2 .loc()标签筛选方法
oly.loc[2]     #显示行标签为2的那一行
oly.loc[2:6]     #显示行标签为2，3,4,5,6那几行,包括行标签为6那行
oly.loc[[2,3,4,5]]  #同上一行一样
oly.loc[[2,4],['City','Sport','Event']] #按行标签和列标签选择
#3 .iloc()按行号和列号选择
oly.iloc[2]     #显示第二行数据，这个方法也可以加列的序号
oly.iloc[2:7]
oly.iloc[2:7,3:7]  #显示第2行到第6行，第3列到第6列的数据
oly.iloc[[1,3,5],[2,4,6]]
```

#### (2)布尔值索引筛选（重点掌握）

```python
#1.利用布尔索引筛选数据
oly.Medal=='Gold'
oly[oly.Medal=='Gold']  #筛选出拿到金牌的那些数据行，生成新的DataFrame
oly1=oly.loc[oly.Medal=='Gold','Event']             #这个非常有用
oly2=oly.loc[oly.Medal=='Gold',['Event','Sport']]   #这个非常有用
oly3=oly[(oly.Medal=='Gold')&(oly.Gender=='Women')]  #生成新的DataFrame
print(oly2.head())
print('-----------------------------------------------------------------')
print(oly3.head(6))
```

#### (3)字符串值索引筛选

```python
#字符串索引筛选数据
#Series有str属性，str属性有contains方法
#oly.Athlete是个Series
#查找特征名称Athlete（运动员）那一列名字叫Florenc的布尔值
bool_value=oly.Athlete.str.contains('Florenc') #返回一个布尔值Series
oly[bool_value] #显示运动员名字叫Florenc那些行
```

```python
#参考答案
import pandas as pd
my_data=pd.read_excel('sales-funnel.xlsx')
#(1)
my_data1=my_data.sort_values(by='Price',ascending=True)
my_data1.head()
#(2)
my_data2=my_data.sort_index(axis=0,ascending=True)
my_data2.head()
my_data3=my_data.sort_index(axis=1,ascending=True)
my_data3.head()
```
