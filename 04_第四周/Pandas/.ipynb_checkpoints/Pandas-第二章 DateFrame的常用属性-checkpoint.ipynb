## 第二章 DateFrame的常用属性

### 1.DataFrame的重要属性

```python
import pandas as pd
import numpy as np
oly=pd.read_csv('olympics.csv',skiprows=4)
oly.shape    #和array一样，是DataFrame的形状，返回一个元组。oly.shape[0]就是数据的行数，oly.shape[1]就是数据的列数
oly.size     #和array一样，是元素总个数
oly.ndim     #和array一样，是该DataFrame的维度
oly.dtypes    #该DataFrame各列的数据类型（元素之间类型可能不一致，但各列的元素类型是一样的）
oly.columns  #是DataFrame的列索引，可以看成一个列表并进行相应的操作
oly.index    #是DataFrame的行索引，可以看成一个列表并进行相应的操作
oly.values   #DataFrame的值，是一个二维数组,一个具有特殊数据类型的二维数组，元素类型不一致。
oly.T   #做转置，行列互换
```

### 2.DataFrame中属性index、columns、values的修改问题

```python
import pandas as pd
import numpy as np
#（1）通过修改f5.columns,f5.index就可以修改f5这个DataFrame的行、列索引
data2=np.arange(12).reshape((3,4))
f5=pd.DataFrame(data2,columns=['n1','n2','n3','n4'],index=['wyx','yyf','glf'])
print(f5)
print()
f5.columns=['xn1','xn2','xn3','xn4']  #可以直接修改DataFrame的列索引
f5.index=['xwyx','xyyf','xglf']       #可以直接修改DataFrame的行索引
print(f5)
#(2) 不能直接给f5.values赋值修改，只能通过f5.values的切片赋值来修改
#f5.values=np.random.random((3,4))   #这个会报错
f5.values[:,1]=100  #这个切片方式对f5.values中的数据类型不一致也可以
```

### 3.DataFrame的常用方法head()、tail()、info()

```python
import numpy as np
import pandas as pd
oly=pd.read_csv('olympics.csv',skiprows=4)
oly.head(4)    #此方法显示前4行，是方法不是属性
oly.tail(6)    #此方法显示后6行
oly.info()   #此方法简单描述数据的特征，显示缺失值数据个数以及内存占用情况
```
