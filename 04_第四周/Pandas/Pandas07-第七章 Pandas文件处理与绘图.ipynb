### 2.利用Series和DataFrame的方法绘图

#### (1)Series的方法绘图

```python
#Series方法绘图(只有一列数据，所无法画散点图，画连线图时横坐标自动设为[0,1,...,])
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
my_series=pd.Series(np.array([2.5,4.1,2.7,8.8,1.0])) #图形的X轴坐标就是Series的索引值。
my_series.plot()   
plt.show()
#也可以按如下方法直接调用plt.plot()画series的图形。
plt.plot(my_series)
plt.show()
```

#### (2)DataFrame的方法绘图

* #### 散点图

```python
#画出DataFrame中任意两列Series构成的坐标的散点图
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
# 产生1000个4列正态分布的随机数
np.random.seed(20)
data = pd.DataFrame(np.random.randn(1000,4),columns=list('ABCD'))
data = data.cumsum()  #对其中的数据进行累加，使例子中显示的曲线有上升波动效果
print(data.head())

#方法1：利用dataframe的方法画任意两列散点图
picture1=data.plot(x='A',y='B',kind='scatter',c='red')  #画出data['A']和data['B']构成的散点图
#picture1=data.plot.scatter(x='A',y='B',c='red') #上边也可以用这行代码实现 
data.plot(x='A',y='C',kind='scatter',c='blue',ax=picture1)  #和第一张图画在一起
plt.show()

#方法2：直接利用plt.scatter()画任意两列的散点图
plt.scatter(data['A'],data['B'],c='r')
plt.scatter(data['A'],data['C'],c='b')
plt.show()
```

* #### 连线图（数据分析前常用）

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
# 产生1000个4列正态分布的随机数
np.random.seed(20)
data = pd.DataFrame(np.random.randn(1000,4),columns=list('ABCD'))
data = data.cumsum() #对每一列数据进行累加，使例子中显示的曲线有上升波动效果
print(data.head())  #打印出前几行的数据，方便我们调试其中的数据值

#1.画出DataFrame中每一列作为纵坐标的图形，横坐标统一为DataFrame的index
data.plot()  #画出DataFrame中每一列作为纵坐标的图形，横坐标统一为DataFrame的index
#plt.plot(data)   #也可以直接调用plt.plot()来画每一列的图形
plt.show()

#2.画出DataFrame中列标签为'A'和'D'的两列构成的点的连线图
data.plot(x='A',y='D',kind='line')  #kind默认为'line',即画连线图。
plt.show()
```

* #### 频率直方图和密度图

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
sampleNo = 500;   #样本个数
mu = 50            #均值
sigma = 10         #标准差

#1.Series的频率直方图和密度图
my_array1 = np.random.normal(mu, sigma, sampleNo)
my_series=pd.Series(my_array1)   #生成一个series
my_series.plot(bins=30,kind='hist',normed=True)  #纵坐标是每个间隔中频率。normed=False表示纵坐标是每个间隔中包含的值的个数
#上边一行也可以写为：my_series.plot.hist(bins=30,normed=True)
my_series.plot(kind='density')
#上边一行也可以写为：my_series.plot.density()
plt.show()

#2.DataFrame的各列的频率直方图和密度图
sampleNo = 500;   #样本个数
mu = 50            #均值
sigma = 10         #标准差
my_array2 = np.random.normal(mu, sigma, size=(sampleNo,4))
my_data=pd.DataFrame(my_array2,columns=list('ABCD'))
picture1=my_data.plot(bins=10,kind='hist',normed=True)
my_data.plot(kind='density',ax=picture1)  #不加ax参数的话密度图和频率直方图不能画在同一张图上。
plt.show()
```

* #### 数据科学分析常用的另一种图像显示(调用seaborn包中的函数实现，非常有用，很重要)

```python
#掌握这个内容，很重要
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sn

#1.读入数据
#此数据文件无列名，因此需要设置header=None，names=['name1','name2','name3','name4']
iris_data = pd.read_csv('iris.data',header=None,names=['sepal_length_cm','sepal_width_cm','petal_length_cm','petal_width_cm','class'])  #这个数据文件人为加了特征名字
iris_data.head()

#2.给出各数据列的统计特征表
print(iris_data.describe())
print('--------------------------------------------------------------------')

#3.数据可视化分析各特征列的特征
##画出按类各数据列之间散点图以及各数据列的频率直方图（各列数据分类显示）。
#主要目的可以看一看能不能根据这些特征图来区分类别。
#若类别那一列是数字型数据，也将和各列一样参与相互关系显示图。
sn.pairplot(iris_data.dropna(), hue='class') #hue='class'表示按照iris_data.class值的不同画不同的颜色
plt.show()
print('--------------------------------------------------------------------')

#4.画出各特征列的数据分布小提琴图
plt.figure(figsize=(10, 10))
for column_index, column in enumerate(iris_data.columns):
    if column == 'class':
        continue
    plt.subplot(2, 2, column_index + 1)
    sn.violinplot(x='class', y=column, data=iris_data)
plt.show()
```

#### 综合练习题：读入数据文件HR_left.csv，该文件是一个职工离职情况数据表，该文件共有5列数据，最后一列标签是'left'，left=1表示该员工已离职，left=0表示该员工未离职。  
(1)将第2列数据的频率直方图和密度图画在同一张图版中；      
(2)画出第1列数据和第2列数据构成的散点图;     
(3)画出第3列数据和第4列数据构成的点的连线图;  
(4)给出个数据列的基本统计描述；  
(5)给出按类各列数据的小提琴图；  
(6)利用sn.pairplot()函数给出各列数据按类的相互关系散点图。

```python
##### 参考答案：
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sn
my_data=pd.read_csv('HR_left.csv')

#第(1)个问题答案
my_data.last_evaluation.plot(bins=30,kind='hist',normed=True)
my_data.last_evaluation.plot(kind='density')
#plt.show()

#第(2)个问题答案
my_data.plot(x='satisfaction_level',y='last_evaluation',kind='scatter')
#plt.show()

#第(3)个问题答案
my_data.plot(x='number_project',y='time_spend_company',kind='line')
#plt.show()

#第（4）个问题答案
my_data.describe()

#第（5）个问题答案
plt.figure(figsize=(10, 10))
for column_index, column in enumerate(my_data.columns):
    if column == 'left':
        continue
    plt.subplot(2, 2, column_index + 1)
    sn.violinplot(x='left', y=column, data=my_data)

#第（6）个问题的答案
sn.pairplot(my_data,hue='left')
plt.show()
```
