#### 绘图

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
plt.rcParams['font.sans-serif'] = ['SimHei']
```

```python

def f(t):
    return np.exp(-t) * np.cos(2*np.pi*t)

a = np.arange(0.0, 5.0, 0.02)

plt.subplot(211)
plt.plot(a, f(a))

plt.subplot(212)
plt.plot(a,np.cos(2*np.pi*2*a), 'r--')
plt.show()
```

```python
a = np.arange(10)
plt.plot(a, a*1.5, 'go-',a, a*2.5, 'rx', a, a*3.5, '*', a, a*4.5, 'b-.')
plt.show()
```

#### python绘图的基本流程
1. 导入库
2. 准备数据
3. 创建画布和子图
4. 绘制图形（折线图、柱状图等）
5. 自定义样式（标题、标签、颜色等）
6. 显示或保存图形

```python
# 不带子图的绘制

# 1. 准备数据
x = np.linspace(0, 10, 100)  # 0到10的100个点
y = np.sin(x)

# 2. 创建画布和子图
plt.figure(figsize=(8, 4))  # 画布大小

# 3. 绘制图形
plt.plot(x, y, label='sin(x)', color='blue', linewidth=2)

# 4. 自定义样式
plt.title("Sine Wave", fontsize=14)
plt.xlabel("X-axis", fontsize=12)
plt.ylabel("Y-axis", fontsize=12)
plt.grid(True, linestyle='--', alpha=0.6)  # 网格线
plt.legend()  # 显示图例

# 5. 显示或保存
plt.tight_layout()  # 自动调整布局
plt.savefig("sine_wave.png", dpi=300)  # 保存为图片
plt.show()  # 显示图形
```

```python
# 带有子图的绘制

# 1. 准备数据
x = np.arange(1, 6)
y1 = np.array([10, 20, 15, 25, 30])
y2 = np.array([5, 15, 10, 20, 25])

# 2. 创建画布和子图（1行2列）
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 4))

# 3. 子图1：柱状图
ax1.bar(x, y1, color='skyblue', label='Group A')
ax1.set_title("Bar Chart")
ax1.set_xlabel("Category")
ax1.set_ylabel("Value")

# 4. 子图2：折线图
ax2.plot(x, y2, marker='o', color='orange', label='Group B')
ax2.set_title("Line Chart")
ax2.set_xlabel("Category")

# 5. 全局设置
fig.suptitle("Combined Plots", fontsize=16)
plt.tight_layout()
plt.show()
```

```python
gender_counts = df['Sex'].value_counts()
plt.pie(gender_counts, labels=['男性','女性'], autopct='%1.1f%%')
plt.title('Gender Distribution')
plt.show()
```

```python
casualty = df['Survived'].value_counts()
plt.pie(casualty, labels=['Died', 'Survived'], autopct='%1.1f%%', startangle=90)
plt.title('Survival Rate')
plt.show()
```

```python
survived_counts = df[df['Survived'] == 1]['Sex'].value_counts()
plt.pie(survived_counts, labels=['女性', '男性'], autopct='%1.1f%%')
plt.title('Survived by Sex')
plt.show()
```

```python
# 不同船票等级幸存者比例饼图
survived_by_Pclass_counts = df[df['Survived'] == 1]['Pclass'].value_counts()
plt.pie(survived_by_Pclass_counts, labels=['1st Class', '2nd Class', '3rd Class'], autopct='%1.1f%%')
plt.show()
```

```python
survived_by_Pclass_counts
```

```python
# 不同等级的仓室下,幸存者中那女比例
c1 = df[(df['Survived'] == 1) & (df['Pclass'] == 1)]['Sex'].value_counts()
c2 = df[(df['Survived'] == 1) & (df['Pclass'] == 2)]['Sex'].value_counts()
c3 = df[(df['Survived'] == 1) & (df['Pclass'] == 3)]['Sex'].value_counts()

fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(10, 5))

ax1.pie(c1, labels=['女', '男'], autopct='%1.1f%%', startangle=90, colors=['pink', 'lightblue'])
ax1.set_title('仓室1')

ax2.pie(c2, labels=['女', '男'], autopct='%1.1f%%', startangle=90, colors=['pink', 'lightblue'])
ax2.set_title('仓室2')
ax3.pie(c3, labels=['女', '男'], autopct='%1.1f%%', startangle=90, colors=['pink', 'lightblue'])
ax3.set_title('仓室3')

plt.show()

```
