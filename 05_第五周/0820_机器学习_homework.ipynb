#### PAT-1088

```python
def main():
    data = input().split()
    M = int(data[0])
    X = int(data[1])
    Y = int(data[2])
    
    a = None
    b = None
    c = None
    
    for i in range(99, 9, -1):
        A = i
        strA = str(A)
        if len(strA) < 2:
            strA = '0' + strA
        B = int(strA[1] + strA[0])
        diff = abs(A - B)
        C_val = B / Y
        if abs(diff - C_val * X) < 1e-5:
            a = A
            b = B
            c = C_val
            break
            
    if a is None:
        print("No Solution")
    else:
        res = [str(a)]
        for person in [a, b, c]:
            if person > M:
                res.append("Cong")
            elif person < M:
                res.append("Gai")
            else:
                res.append("Ping")
        print(" ".join(res))

if __name__ == "__main__":
    main()
```

#### PAT-1089

```python
def main():
    n = int(input().strip())
    statements = []
    for _ in range(n):
        s = input().strip()
        num = int(s)
        statements.append(num)
    
    # 遍历所有可能的狼人组合（i和j，且i<j）
    for i in range(1, n):
        for j in range(i+1, n+1):
            # 假设i和j是狼人，其余是好人
            # 创建身份数组：1表示好人，-1表示狼人
            identity = [1] * (n+1)
            identity[i] = -1
            identity[j] = -1
            
            liars = []  # 记录说谎的人
            for k in range(1, n+1):
                said = statements[k-1]
                # 玩家k说的人物编号是abs(said)，身份符号与said相同（正为好，负为狼）
                target = abs(said)
                # 实际身份是identity[target]
                if said > 0:
                    # 说target是好人
                    if identity[target] != 1:
                        liars.append(k)
                else:
                    # 说target是狼人
                    if identity[target] != -1:
                        liars.append(k)
            
            # 必须恰好有2人说谎
            if len(liars) != 2:
                continue
                
            # 检查说谎的两人中是否至少有一个狼人（但并非所有狼人都说谎）
            wolf_liars = 0
            for liar in liars:
                if identity[liar] == -1:
                    wolf_liars += 1
            
            if wolf_liars == 1:  # 恰好一个狼人说谎（另一个狼人说真话）
                print(f"{i} {j}")
                return
                
    print("No Solution")

if __name__ == "__main__":
    main()
```

#### PAT-1090

```python
def main():
    import sys
    data = sys.stdin.read().splitlines()
    n, m = map(int, data[0].split())
    incompatible = {}
    
    index = 1
    for i in range(n):
        a, b = data[index].split()
        index += 1
        if a not in incompatible:
            incompatible[a] = set()
        incompatible[a].add(b)
        if b not in incompatible:
            incompatible[b] = set()
        incompatible[b].add(a)
    
    results = []
    for i in range(m):
        parts = data[index].split()
        index += 1
        k = int(parts[0])
        items = parts[1:1+k]
        item_set = set(items)
        safe = True
        for item in items:
            if item in incompatible:
                for bad in incompatible[item]:
                    if bad in item_set:
                        safe = False
                        break
                if not safe:
                    break
        results.append("Yes" if safe else "No")
    
    for res in results:
        print(res)

if __name__ == "__main__":
    main()
```
